```python
# Precision Plot
# width value of the bar plots
width = .1
# create figure
fig = plt.subplots(figsize =(16, 8)) 

# Set position of bar on X axis 
b1 = np.arange(len(stages)) 
b2 = [x + width for x in b1] 
b3 = [x + width for x in b2]
b4 = [x + width for x in b3]
b5 = [x + width for x in b4]
b6 = [x + width for x in b5]

# Plot each Sleep Stage based on the Model it is grouped with
# Sleep Stage 1
plt.bar(b1, p1, color ='red', width = width, edgecolor ='grey', label ='Sleep Stage 1')
# Sleep Stage 2
plt.bar(b2, p2, color ='green', width = width, edgecolor ='grey', label ='Sleep Stage 2')
# Sleep Stage 3
plt.bar(b3, p3, color ='blue', width = width, edgecolor ='grey', label ='Sleep Stage 3')
# Sleep Stage 4
plt.bar(b4, p4, color ='purple', width = width, edgecolor ='grey', label ='Sleep Stage 4')
# Sleep Stage Rem
plt.bar(b5, p5, color ='pink', width = width, edgecolor ='grey', label ='Sleep Stage R') 
# Movement time
plt.bar(b6, p6, color ='orange', width = width, edgecolor ='grey', label ='Movement time') 

# X label
plt.xlabel('Model', fontweight ='bold', fontsize = 15)
# Y Label
plt.ylabel('Precision', fontweight ='bold', fontsize = 15)

plt.xticks([r + width for r in range(len(stages))],['Random Forest Classifier','Random Forest Classifier (HPT)','Decision Tree','Decision Tree (HPT)','XGBoost','XGBoost (HPT)'])

plt.legend()
plt.show()
```

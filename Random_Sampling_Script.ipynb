```python
import os
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
# replace path with path of stored files
path = "C:\\Users\\black\\OneDrive\\Desktop\\Sleep Project Files"

# number of patients to use
count = 10 
files = []

for i in os.listdir(path):
    if count > 0:
        files.append(i)
        count=count-1
    else:
        break
```

```python
print(files)
```

```python
# percentage of samples to use
sample_size = 0.6

mixed_df = pd.DataFrame()

for i in files:
    file_path = path+'\\'+i
    #print(file_path)
    
    df = pd.read_csv(file_path,low_memory=False)
    
    if 'Movement time' in df['target'].values:
        df = df.replace('Movement time',6)
    
    #Proportionate Stratified Random Sampling
    df = df.groupby('target',group_keys=False).apply(lambda x: x.sample(frac=sample_size))
    mixed_df = pd.concat([mixed_df,df])
    
```

```python
X = mixed_df.drop('target', axis = 1)
display(X)
```
